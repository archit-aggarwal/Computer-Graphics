### Mid Point Circle Drawing Algorithm

```python
def PutPixle(win, x, y):
    pt = Point(x,y)
    pt.draw(win)        
    # win.plotPixel(x_coords[i], y_coords[i], "Blue")
```

```python
def midPointCircleDraw():
    x_centre = int(input("Enter X coordinate of center: "))
    y_centre = int(input("Enter Y coordinate of center: "))
    r = int(input("Enter radius of circle: "))
    
    win = GraphWin('Mid Point Circle', 600, 480)
    #initial values
    x = 0
    y = r
    
    oct_1 = []
    oct_2 = []
    oct_1.append([x,y])
    oct_2.insert(0, [y,x])

    P = 1 - r 
  
    while x<y:
        x += 1
        
        if P<0: 
            P = P + 2*x + 1
        else:         
            y -= 1
            P = P - 2*y + 2*x + 1
        
        if (x > y):
            break
            
        oct_1.append([x,y])
        if x != y:
            oct_2.insert(0, [y,x])
        
    quad_1 = oct_1 + oct_2
    quad_2 = []
    quad_3 = []
    quad_4 = []
    
    for (x,y) in quad_1:
        quad_2.append([-x,y])
        quad_3.append([-x,-y])
        quad_4.append([x,-y])
    
    print("Quadrant 1", "Quadrant 2", "Quadrant 3", "Quadrant 4")
    for i in range(len(quad_1)):
        quad_1[i][0] += x_centre
        quad_1[i][1] += y_centre
        quad_2[i][0] += x_centre
        quad_2[i][1] += y_centre
        quad_3[i][0] += x_centre
        quad_3[i][1] += y_centre
        quad_4[i][0] += x_centre
        quad_4[i][1] += y_centre
        
        print(quad_1[i], quad_2[i], quad_3[i], quad_4[i])
        
        PutPixle(win, quad_1[i][0], quad_1[i][1])
        PutPixle(win, quad_2[i][0], quad_2[i][1])
        PutPixle(win, quad_3[i][0], quad_3[i][1])
        PutPixle(win, quad_4[i][0], quad_4[i][1])
            
    
    # win.getMouse()
    win.close()           
```

### Mid Point Ellipse Drawing Algorithm

```python
def midPointEllipseDraw():
    xc = int(input("Enter X coordinate of center: "))
    yc = int(input("Enter Y coordinate of center: "))
    rx = int(input("Enter horizontal radius: "))
    ry = int(input("Enter vertical radius: "))
    
    x = 0;
    y = ry;
        
    win = GraphWin('Mid Point Ellipse', 600, 480)
    
    # Initial decision parameter of region 1
    d1 = ((ry * ry) - (rx * rx * ry) + (0.25 * rx * rx));
    dx = 2 * ry * ry * x;
    dy = 2 * rx * rx * y;

    # For region 1
    while (dx < dy):
        # Print points based on 4-way symmetry
        x1 = x+xc
        y1 = y+yc
        x2 = -x+xc
        y2 = y+yc
        x3 = -x+xc
        y3 = -y+yc
        x4 = x+xc
        y4 = -y+yc
        
        print("(", x1, ",", y1, ") ", sep="", end="")
        print("(", x2, ",", y2, ") ", sep="", end="")
        print("(", x3, ",", y3, ") ", sep="", end="")
        print("(", x4, ",", y4, ") ", sep="")
        PutPixle(win, x1, y1)
        PutPixle(win, x2, y2)
        PutPixle(win, x3, y3)
        PutPixle(win, x4, y4)

        # Checking and updating value of decision parameter based on algorithm
        if (d1 < 0):
            x += 1;
            dx = dx + (2 * ry * ry);
            d1 = d1 + dx + (ry * ry);
        else:
            x += 1;
            y -= 1;
            dx = dx + (2 * ry * ry);
            dy = dy - (2 * rx * rx);
            d1 = d1 + dx - dy + (ry * ry);

    # Decision parameter of region 2
    d2 = (((ry * ry) * ((x + 0.5) * (x + 0.5))) + ((rx * rx) * ((y - 1) * (y - 1))) -
        (rx * rx * ry * ry));

    # Plotting points of region 2
    while (y >= 0):
        # printing points based on 4-way symmetry
        x1 = x+xc
        y1 = y+yc
        x2 = -x+xc
        y2 = y+yc
        x3 = -x+xc
        y3 = -y+yc
        x4 = x+xc
        y4 = -y+yc
        
        print("(", x1, ",", y1, ") ", sep="", end="")
        print("(", x2, ",", y2, ") ", sep="", end="")
        print("(", x3, ",", y3, ") ", sep="", end="")
        print("(", x4, ",", y4, ") ", sep="")
        PutPixle(win, x1, y1)
        PutPixle(win, x2, y2)
        PutPixle(win, x3, y3)
        PutPixle(win, x4, y4)

        # Checking and updating decision parameter value based on algorithm
        if (d2 > 0):
            y -= 1;
            dy = dy - (2 * rx * rx);
            d2 = d2 + (rx * rx) - dy;
        else:
            y -= 1;
            x += 1;
            dx = dx + (2 * ry * ry);
            dy = dy - (2 * rx * rx);
            d2 = d2 + dx - dy + (rx * rx);
        
    win.getMouse()
    win.close()    
```
